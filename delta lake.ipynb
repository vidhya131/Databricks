---

# 🏗️ 1️⃣ Managed Table

## 🔹 What It Means

A **Managed Table** is fully controlled by Databricks.

* Databricks manages:

  * Storage location
  * Metadata
  * File lifecycle

If you **drop the table**, the data files are also deleted.

---

## 🔹 Example

```sql
CREATE TABLE sales (
  id INT,
  amount DOUBLE
);
```

👉 No `LOCATION` specified
👉 Automatically becomes **managed**

---

## 🔹 Where Is Data Stored?

### Hive Metastore:

* Stored in DBFS default warehouse location

### Unity Catalog:

* Stored in catalog’s managed storage location

You don’t control the exact path.

---

## 🔹 What Happens on DROP?

```sql
DROP TABLE sales;
```

✔ Metadata deleted
✔ Data deleted

---

## 🔹 When Industry Uses Managed Tables

* Silver layer tables
* Gold aggregated tables
* Internal transformation outputs
* Temporary analytical tables

Basically:
👉 Data that is owned and controlled by Databricks.

---

# 🌍 2️⃣ External Table

## 🔹 What It Means

An **External Table**:

* Metadata stored in Databricks
* Data stored in a user-defined location (ADLS/S3)
* Databricks does NOT own the storage

If you drop the table → data remains.

---

## 🔹 Example

```sql
CREATE TABLE ext_sales
USING DELTA
LOCATION 'abfss://bronze@storageaccount.dfs.core.windows.net/sales';
```

👉 LOCATION specified
👉 Becomes **external**

---

## 🔹 What Happens on DROP?

```sql
DROP TABLE ext_sales;
```

✔ Metadata deleted
❌ Data remains in ADLS

---

## 🔹 When Industry Uses External Tables

* Bronze/raw data layer
* Shared enterprise data lake
* Data shared across tools (Synapse, Snowflake, etc.)
* Regulatory data that must not be accidentally deleted

---

# ⚖️ Managed vs External — Clear Comparison

| Feature                     | Managed Table            | External Table                     |
| --------------------------- | ------------------------ | ---------------------------------- |
| Storage location            | Controlled by Databricks | User-defined                       |
| Drop deletes data?          | ✅ Yes                    | ❌ No                               |
| Best for                    | Internal pipelines       | Shared data lake                   |
| Risk of accidental deletion | Higher                   | Lower                              |
| Storage governance          | Automatic                | Needs external location setup (UC) |

---

# 🏢 Real Industry Architecture Example

## 🥉 Bronze Layer

External tables
Reason:

* Raw data must never be deleted
* Owned by enterprise storage team

---

## 🥈 Silver Layer

Managed tables
Reason:

* Derived data
* Controlled lifecycle

---

## 🥇 Gold Layer

Managed tables
Reason:

* Analytics-ready
* Fully governed in Databricks

---

# 🧠 In Unity Catalog Context

For external tables, you must create:

1. Storage Credential
2. External Location
3. Then create table

Managed tables don’t need that — they use catalog-managed storage.

---

# 🎯 Interview-Level Answer

> Managed tables are fully controlled by Databricks and their data is deleted when dropped. External tables store metadata in Databricks but keep data in user-managed storage, and dropping them does not delete the underlying files.

---

# 🚨 Common Mistake in Industry

People accidentally:

```sql
DROP TABLE managed_table;
```

And lose data permanently.

That’s why:

* Bronze is usually external
* Silver/Gold are managed

---

# 🏁 Final Summary

## Managed Table

* Databricks owns storage
* Drop = data gone

## External Table

* You own storage
* Drop = metadata only removed
